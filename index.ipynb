#### Write a function density() that takes in a random variable and calculates the density function using `np.hist` and interpolation. The function should return two lists carrying x and y coordinates for plotting the density function

```python
def density(x):
    
    n, bins = np.histogram(x, 10, density=1)
    # Initialize numpy arrays with zeros to store interpolated values
    pdfx = np.zeros(n.size)
    pdfy = np.zeros(n.size)

    # Interpolate through histogram bins 
    # identify middle point between two neighbouring bins, in terms of x and y coords
    for k in range(n.size):
        pdfx[k] = 0.5*(bins[k]+bins[k+1])
        pdfy[k] = n[k]

    # plot the calculated curve
    return pdfx, pdfy


# Generate test data and test the function
np.random.seed(5)
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 100)
x,y = density(s)
plt.plot(x,y, label = 'test')
plt.legend()
plt.show()
```

#### Add Overlapping density plots for male and female heights to the histograms plotted earlier

```python
# You code here 
a,b = density(male_df.Height)
c,d = density(female_df.Height)
plt.figure(figsize=(12,8))
plt.hist(male_df.Height, bins=10, color='red', alpha=.5, label='Male Height', normed=True)
plt.hist(female_df.Height, bins=10, color='blue', alpha=.5, label='Female Height', normed=True)
plt.plot(a,b, label = 'Male Height Density Plot')
plt.plot(c,d, label = 'Female Height Density Plot')
plt.title('Male and Female Heights Histogram')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.legend();
```

#### Repeat above exerice for male and female weights

```python
# Your code here 
e,f = density(male_df.Weight)
g,h = density(female_df.Weight)
plt.figure(figsize=(12,8))
plt.hist(male_df.Weight, bins=10, color='red', alpha=.5, label='Male Weight', normed=True)
plt.hist(female_df.Weight, bins=10, color='blue', alpha=.5, label='Female Weight', normed=True)
plt.plot(e,f, label = 'Male Weight Density Plot')
plt.plot(g,h, label = 'Female Weight Density Plot')
plt.title('Male and Female Weights Histogram')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.legend();
```
